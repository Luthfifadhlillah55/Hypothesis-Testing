## Confidence Interval
a confidence interval (CI) is a range of estimates for an unknown parameter, defined as an interval with a lower bound and an upper bound (notwithstanding one-sided confidence intervals, which are bounded only on one side). 

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

goal = df1['goal']
m = goal.mean() #MEAN value of the goal column.
s = goal.std()  #Standard deviation value of the goal column.
c = stats.norm.interval(0.95, m, s) #ci value

goal.plot.hist(title='Sample Distribution', bins=100, density=True) 
plt.axvline(m, color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(c[0], color='green', linestyle='dashed', linewidth=2)
plt.axvline(c[1], color='green', linestyle='dashed', linewidth=2)
plt.show()

print ('MEAN of the Sample: ' + str(m))
print ('STD of the Sample: ' + str(s))
print ('Confidence Interval (95%): ' + str(c))
```

```python
backer = df1['backers']
m = backer.mean() #MEAN value of the backers column.
s = backer.std()  #STD value of the backers column.
c = stats.norm.interval(0.95, m, s) #ci value

backer.plot.hist(title='Sample Distribusi', bins=100, density=True) 
plt.axvline(m, color='yellow', linestyle='dashed', linewidth=2)
plt.axvline(c[0], color='green', linestyle='dashed', linewidth=2)
plt.axvline(c[1], color='green', linestyle='dashed', linewidth=2)
plt.show()

print ('MEAN of the Sample: ' + str(m))
print ('STD of the Sample: ' + str(s))
print ('Confidence Interval (95%): ' + str(c))
```

## Hypothesis Testing
Hypothesis testing is an act in statistics whereby an analyst tests an assumption regarding a population parameter. The methodology employed by the analyst depends on the nature of the data used and the reason for the analysis.
<br>

Null hypothesis (H0): A hypothesis associated with a theory one would like to prove. 
<br>
Alternative hypothesis (H1): A hypothesis associated with a contradiction to a theory one would like to prove.
<br>
Critical value: The threshold value of the test statistic for rejecting the null hypothesis

```python
df.head()
```

```python
GB_pledged_real = df[df['country']=='GB'][['usd_pledged_real','pledged']].groupby('pledged').sum()
US_pledged_real = df[df['country']=='US'][['usd_pledged_real','pledged']].groupby('pledged').sum()

print('Average usd_pledged_real of GB: ${}'.format(np.round(GB_pledged_real.usd_pledged_real.mean())))
print('Average usd_pledged_real of US : ${}'.format(np.round(US_pledged_real.usd_pledged_real.mean())))
```

The hypothesis on this case:

**H0: Î¼_GB = Î¼_US**

**H1: Î¼_GB != Î¼_US**

```python
t_stat, p_val = stats.ttest_ind(GB_pledged_real,US_pledged_real)
print('P-value:',p_val[0]) #the p-value isn't divided by 2 since the output is two-sided p-value
print('t-statistics:',t_stat[0])
```
